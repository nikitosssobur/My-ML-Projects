```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import preprocessing as proc
from sklearn.linear_model import LogisticRegression
```

```python
diabet_data = pd.read_csv("Datasets\diabetes2.csv")
diabet_data
```

```python
diabet_data.info()
```

```python

features_data = diabet_data[diabet_data.columns[:-1]]
features_data
```

```python
target_data = diabet_data[diabet_data.columns[-1]]
target_data
```

```python
diabet_data.isnull().any() 
```

```python
#preprocessing data, dividing to train/test samples, converting to numpy arrays, scaling x_data
x_train, x_test, y_train, y_test = train_test_split(features_data, target_data, test_size = 0.33, random_state = 42)
x_train_np_data, x_test_np_data = np.array(x_train), np.array(x_test)
y_train_np_data, y_test_np_data = np.array(y_train), np.array(y_test)
scaler = proc.StandardScaler().fit(x_train_np_data)
x_train_scaled = scaler.transform(x_train_np_data)
x_test_scaled = scaler.transform(x_test_np_data)
```

```python
#creating logistic regression model for doing prediction
logit_model = LogisticRegression(random_state = 0).fit(x_train_scaled, y_train_np_data)

#predicting values by test input
y_pred = logit_model.predict(x_test_scaled)
y_pred
```

```python
test_true_data = y_test_np_data[y_test_np_data == 1]
train_true_data = y_train_np_data[y_train_np_data == 1]
print(sum(test_true_data) / len(y_test_np_data))
print(sum(train_true_data) / len(y_train_np_data))
print((sum(test_true_data) + sum(train_true_data)) / len(target_data))
```

```python
result = y_test_np_data[y_test_np_data == y_pred]
print(len(result) / len(y_test_np_data))
```
